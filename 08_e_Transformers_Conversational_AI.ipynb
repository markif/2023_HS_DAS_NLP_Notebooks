```python
import transformers

conversational_pipeline = transformers.pipeline("conversational")
```

```python
conversation_1 = transformers.Conversation("Going to the movies tonight - any suggestions?")
conversation_2 = transformers.Conversation("What's the last book you have read?")
```

```python
conversational_pipeline([conversation_1, conversation_2])
```

```python
conversation_1.add_user_input("Is it an action movie?")
conversation_2.add_user_input("What is the genre of this book?")
```

```python
conversational_pipeline([conversation_1, conversation_2])
```
